# **Setup**

```python
import pandas as pd 
```

# **Data Import**

```python
df = pd.read_csv('ocd_patient_dataset.csv')

print(df.shape)
df.head()
```

# **Diagnosis MOM Count**

```python
df01 = df.loc[:,['Patient ID','OCD Diagnosis Date']]

df01.head()
```

```python
df01.info()
```

```python
df01.rename(columns={'OCD Diagnosis Date':'OCD_Diagnosis_Date'}, inplace=True)
df01['OCD_Diagnosis_Date'] = pd.to_datetime(df01['OCD_Diagnosis_Date'], format='%Y-%m-%d')

display(df01.head())
df01.info()
```

```python
df01['Diagnosis_Month'] = df01['OCD_Diagnosis_Date'].apply(lambda x: x.strftime('%m-01-%Y'))
df01['Diagnosis_Month'] = pd.to_datetime(df01['Diagnosis_Month'], format='%m-01-%Y')
df01['Diagnosis_Month'] = df01['Diagnosis_Month'].apply(lambda x: x.strftime('%m-01-%Y'))
# df01.reset_index(inplace=True)
df01.sort_values('OCD_Diagnosis_Date',ascending=True,inplace=True)

display(df01.head())
df01.info()
```

```python
df02 = df01.groupby('Diagnosis_Month')['Patient ID'].count().to_frame()
df02.reset_index(inplace=True)
df02.columns = ['Diagnosis_Month','patient_count']

display(df02.head())
df02.info()
```

```python
# Export df01
df02.to_csv('diagnosis_mom.csv', index=False)
```
